# Extract features

```python
def extract_features(xyz_path):
    atoms = read(xyz_path)
    positions = atoms.get_positions()
    charges = atoms.get_atomic_numbers()
    return positions, charges

def load_all_xyz(folder_path, max_atoms=23):
    positions_list = []
    charges_list = []
    
    xyz_files = sorted([
        os.path.join(folder_path, f)
        for f in os.listdir(folder_path)
        if f.endswith('.xyz')
    ])

    for xyz_path in xyz_files:
        pos, chg = extract_features(xyz_path)
        
        pos_padded = np.zeros((max_atoms, 3))
        chg_padded = np.zeros((max_atoms,))
        
        n_atoms = pos.shape[0]
        pos_padded[:n_atoms, :] = pos
        chg_padded[:n_atoms] = chg

        positions_list.append(pos_padded)
        charges_list.append(chg_padded)

    positions_array = np.stack(positions_list)  # shape (N, max_atoms, 3)
    charges_array = np.stack(charges_list)      # shape (N, max_atoms)
    
    return positions_array, charges_array
```

# Some visualization

```python
def visualize_molecule(positions, charges, title="Molecule 3D Structure"):
    # Remove zero-padded atoms (charges==0)
    mask = charges > 0
    positions = positions[mask]
    charges = charges[mask]
    symbols = [chemical_symbols[int(Z)] for Z in charges]

    # Color map for elements (extend as needed)
    color_map = {
        1: 'white',   # H
        6: 'black',   # C
        7: 'blue',    # N
        8: 'red',     # O
        9: 'green',   # F
        16: 'yellow', # S
        17: 'lime',   # Cl
        # Add more if needed
    }
    colors = [color_map.get(int(Z), 'gray') for Z in charges]

    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.set_title(title)

    # Plot atoms
    ax.scatter(positions[:, 0], positions[:, 1], positions[:, 2],
               s=200, c=colors, edgecolor='k', alpha=0.8)

    # Annotate atoms with chemical symbols
    for pos, sym in zip(positions, symbols):
        ax.text(pos[0], pos[1], pos[2], sym, fontsize=12, ha='center', va='center')

    # Equal aspect ratio
    max_range = (positions.max(axis=0) - positions.min(axis=0)).max() / 2.0
    mid = positions.mean(axis=0)
    for i in range(3):
        ax.set_xlim(mid[0] - max_range, mid[0] + max_range)
        ax.set_ylim(mid[1] - max_range, mid[1] + max_range)
        ax.set_zlim(mid[2] - max_range, mid[2] + max_range)

    ax.set_xlabel('X (Å)')
    ax.set_ylabel('Y (Å)')
    ax.set_zlabel('Z (Å)')
    plt.tight_layout()
    plt.show()
```

```python
xyz_files = sorted([f for f in os.listdir(atoms_train_dir) if f.endswith('.xyz')])
first_xyz = os.path.join(atoms_train_dir, xyz_files[0])

positions, charges = extract_features(first_xyz)
visualize_molecule(positions, charges, title=f"3D Structure: {xyz_files[0]}")
```
